```python
import pandas as pd
```

```python
infile = '/home/h05/hadin/Human_habitability/analysis/Habitability_Questionnaire_final_4thJuly2023_translated_a.csv'
df = pd.read_csv(infile, encoding='ISO-8859-1')
```

```python
df
```

```python
df_essential = df.filter(regex='Essential')
df_essential.columns = [x.split('[')[1].rstrip(']') for x in df_essential.columns.to_list()]
df_essential
```

```python
df_essential['Shelter'].value_counts()
```

```python
df_essential.plot.bar(stacked=True)
```

```python

ocols = ['Element', 'Value', 'Count']
odf = pd.DataFrame(columns=ocols)

for col in df_essential.columns:
    vc = df_essential[col].value_counts()
    temp_df = pd.DataFrame({'Element': col, 'Value': vc.index, 'Count': vc.values})
    odf = pd.concat([odf, temp_df])

odf.reset_index(drop=True, inplace=True)
odf
```

```python
odf_wide = pd.pivot_table(odf, values="Count", index="Element", columns=["Value"])
ax = odf_wide.plot.barh(stacked=True)
ax.set_title("What is Essential to Live?",color='black')
ax.legend(bbox_to_anchor=(1.0, 1.0))
```

```python
import seaborn as sns
sns.barplot(x="Count", y="Element", hue="Value", data=odf, orient='h')
```
